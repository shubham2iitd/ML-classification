```python
rows = DF.iloc[:,0].tolist()

collec_str = ""
for i in range(len(rows)):
    if pd.notnull(rows[i]):
        collec_str += " " + rows[i]
    
```

```python
import nltk
from nltk.tokenize import wordpunct_tokenize
```

```python
import re, collections

def words(text):
    tokens = nltk.wordpunct_tokenize(text.lower())
        
    return re.findall('[a-z]+', text.lower()) 

def train(features):
    model = collections.defaultdict(lambda: 1)
    for f in features:
        model[f] += 1
    return model

NWORDS = train(words(collec_str))

alphabet = 'abcdefghijklmnopqrstuvwxyz'

def edits1(word):
   splits     = [(word[:i], word[i:]) for i in range(len(word) + 1)]
   deletes    = [a + b[1:] for a, b in splits if b]
   transposes = [a + b[1] + b[0] + b[2:] for a, b in splits if len(b)>1]
   replaces   = [a + c + b[1:] for a, b in splits for c in alphabet if b]
   inserts    = [a + c + b     for a, b in splits for c in alphabet]
   return set(deletes + transposes + replaces + inserts)

def known_edits2(word):
    return set(e2 for e1 in edits1(word) for e2 in edits1(e1) if e2 in NWORDS)

def known(words): return set(w for w in words if w in NWORDS)

def correct(word):
    candidates = known([word]) or known(edits1(word)) or known_edits2(word) or [word]
    return max(candidates, key=NWORDS.get)
```

```python
df = pd.read_csv("C:\Users\shubham.akshat\Desktop\BODM/TELEGRAPH__RMA__BODM.txt", sep = "\t")

df["free_text"] = df.PROBLEM_DESCRIPTION.str.lower()
df = df[df["HDX_CODEC"] == True]

df = df[pd.notnull(df['free_text'])]
df = df.drop_duplicates(cols = 'free_text')
df.index = range(len(df))
```

```python
all_words = []
row_words = []
for i in range(len(df)):
    row_words.append(nltk.wordpunct_tokenize(df["free_text"][i]))
    
```

```python
all_words = [item for sublist in row_words for item in sublist]
```

```python
from collections import OrderedDict
all_words = OrderedDict.fromkeys(all_words).keys()

print len(all_words)

all_words_fil = []

flag = {}
for word in all_words:
    if any(char.isdigit() for char in word) or len(word) < 3:
        flag[word] = 0
    else:
        flag[word] = 1
for word, flg in flag.items():
    if flg == 1:
        all_words_fil.append(word)


len(all_words_fil)        
```

```python
dff = pd.DataFrame(index = range(len(all_words)))
dff["raw"] = ""
dff["code_correction"] = ""

for i in range(len(all_words_fil)):
    dff["raw"][i] = all_words_fil[i]
    dff["code_correction"][i] = correct(all_words_fil[i])
```

```python
outfile = "C:/Users/shubham.akshat/Desktop/BODM//tfidf//spell_code_output.txt"
dff.to_csv(outfile, sep='\t', index=False)
```
