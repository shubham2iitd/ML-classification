```python
from nltk.corpus import wordnet as wn
import itertools
nltk.tag._POS_TAGGER

def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wn.ADJ
    elif treebank_tag.startswith('V'):
        return wn.VERB
    elif treebank_tag.startswith('N'):
        return wn.NOUN
    elif treebank_tag.startswith('R'):
        return wn.ADV
    else:
        return wn.NOUN


def get_first_synonym(lemma0, pos):
    synonyms = []
    if len(wn.synsets(lemma0, pos)) > 0:
        for synset in wn.synsets(lemma0, pos):
            synonyms.append(synset.lemma_names)
            merged = list(itertools.chain(*synonyms))
            merged.sort()
        return merged[0].lower()
    else:
        return lemma0.lower()
    
```

```python
def stem_tokens(tokens, stemmer):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    return stemmed

def lemma_tokens(tokens, lmtzr):
    lemma_lst = []
    for word in tokens:
        lexicon = nltk.pos_tag([word])
        POS = get_wordnet_pos(lexicon[0][1])        
        lemma_lst.append(lmtzr.lemmatize(word, POS))
    return lemma_lst

nouns = []
adjectives = []
verbs = []
adverbs = []


def tokenize(text):
    
    tokens = nltk.word_tokenize(text)
   # filtered_tokens = [w for w in tokens if not w in stopwords.words('english')]

    
    lexicons = nltk.pos_tag(tokens)
    lemmatised_tokens = []
    for lexicon in lexicons:
        if lexicon[0].lower() in incorrect_words:
            lexicon = (get_correct_word(lexicon[0]), lexicon[1])
            
        if lexicon[0].lower() not in remove_words and lexicon[0] not in stopwords.words('english'):
            wn_pos = get_wordnet_pos(lexicon[1])
            
            if lexicon[0].lower() in POS_other_than_noun:
                lemma0 = lmtzr.lemmatize(lexicon[0].lower(), wn_pos)
                lemma1 = get_first_synonym(lemma0, wn_pos)
                lemmatised_tokens.append(lemma1)
                
#             if wn_pos == wn.ADJ:
#                 adjectives.append(lexicon[0])
#             if wn_pos == wn.VERB:
#                 verbs.append(lexicon[0])
#             if wn_pos == wn.ADV:
#                 adverbs.append(lexicon[0])
            
            elif wn_pos == wn.NOUN:
#                nouns.append(lexicon[0])
                lemma0 = lmtzr.lemmatize(lexicon[0].lower(), wn_pos)
                lemma1 = get_first_synonym(lemma0, wn_pos)
                lemmatised_tokens.append(lemma1)
    
 #   lemmatised_tokens = [w for w in lemmatised_tokens_1 if not w in remove_words]
    
#    return stemmed_tokens
    return lemmatised_tokens
#    return filtered_tokens

```

```python
def prob_sum_value(lda, topic_no, lst, top_word_lst):
    value = 0
    for i in range(len(top_word_lst)):
        if top_word_lst[i] in lst:
            value += lda.show_topic(int(topic_no), topn=50)[i][0]
    return value
        
```

```python
#  FOR TFIDF


import pickle
import string
from nltk.corpus import wordnet as wn
replace_punctuation = string.maketrans(string.punctuation, ' '*len(string.punctuation))

fModel = open("C:\Users\shubham.akshat\Desktop\BODM\Lemma_synonym_dict_TOT.pkl","rb")
lema_to_raw = pickle.load(fModel)
fModel.close()

for n_topics in range(15,16):
    n_topics = int(n_topics)
    
    fModel = open("C:\Users\shubham.akshat\Desktop\BODM\\tfidf\\LDA_classifiers_WN_noun_tst " + str(n_topics) + " topics.pkl","rb")
    classifier = pickle.load(fModel)
    fModel.close()


    key = str(n_topics)+"_topics"
    lda = classifier.get(key)[0]
    dictionary = classifier.get(key)[1]
    tfidf = classifier.get(key)[2]

    
    dff = pd.DataFrame(index = range(len(df)))
    dff["UID"] = df["UID"]
    dff["PROBLEM_DESCRIPTION"] = df["PROBLEM_DESCRIPTION"] 
    dff["NDF"] = df["NDF"]
    
    dff["Top topic"] = ""
    dff["% portion of top topic"] = ""
    dff["Words from top topic"] = ""
    dff["Top topic's top words"] = ""
    dff["Prob_sum_value of top topic"] = ""
    
    
    dff["2nd Top topic"] = ""
    dff["% portion of 2nd top topic"] = ""
    dff["Words from 2nd top topic"] = ""
    dff["2nd Top topic's top words"] = ""
    dff["Prob_sum_value of 2nd top topic"] = ""
    
    topic_dict = {}
    for i in range(n_topics):
        temp = lda.show_topic(topicid = i, topn = 20)
        terms = []
        for term in temp:
            s = lema_to_raw[term[1]].split(",")[0]
            terms.append(s)
        topic_dict["topic "+ str(i)] = terms

    topic_dict_2 = {}
    for i in range(n_topics):
        temp = lda.show_topic(topicid = i, topn = 50)
        terms = []
        for term in temp:
            s = term[1]
            terms.append(s)
        topic_dict_2["topic "+ str(i)] = terms        

    for i in range(len(dff)):
        temp1 = str(dff["PROBLEM_DESCRIPTION"][i]).translate(replace_punctuation)
        temp2 = temp1.decode('utf-8','ignore')
        temp3 = tokenize(temp2)
        input0 = dictionary.doc2bow(temp3)
        input1 = tfidf[input0]
        d = defaultdict()
        
        for tupl in lda.__getitem__(input1, eps=0.001):
     #   for tupl in lda.__getitem__(input0, eps=0.001):
            topic = tupl[0]
            prob = tupl[1]
            d["topic "+ str(topic)] = prob

        n1 = 0
        n2 = 0
        for word in temp3:
            if word in topic_dict_2[sorted(d, key=d.get, reverse=True)[0]]:
                n1 += 1
            if (len(d) > 1):
                if word in topic_dict_2[sorted(d, key=d.get, reverse=True)[1]]:
                    n2 += 1
            
        dff["Top topic"][i] = sorted(d, key=d.get, reverse=True)[0]
        dff["% portion of top topic"][i] = max(d.iteritems(), key=operator.itemgetter(1))[1]
        dff["Words from top topic"][i] = n1
        dff["Top topic's top words"][i] = topic_dict[sorted(d, key=d.get, reverse=True)[0]]
        dff["Prob_sum_value of top topic"][i] = prob_sum_value(lda, sorted(d, key=d.get, reverse=True)[0][-1], temp3, topic_dict_2[sorted(d, key=d.get, reverse=True)[0]])
        
        
  
        dff["2nd Top topic"][i] = sorted(d, key=d.get, reverse=True)[1]
        dff["% portion of 2nd top topic"][i] = d[sorted(d, key=d.get, reverse=True)[1]]
        dff["Words from 2nd top topic"][i] = n2
        dff["2nd Top topic's top words"][i] = topic_dict[sorted(d, key=d.get, reverse=True)[1]]
        dff["Prob_sum_value of 2nd top topic"][i] = prob_sum_value(lda, sorted(d, key=d.get, reverse=True)[1][-1], temp3, topic_dict_2[sorted(d, key=d.get, reverse=True)[1]])
    
    

    outfile = "C:/Users/shubham.akshat/Desktop/BODM//tfidf//topic_modeling_WN_noun_tst2 " + str(n_topics) + "_topics.txt"
    dff.to_csv(outfile, sep='\t', index=False)
```
