```python
df = df.applymap(lambda cell_val: 'BLANK' if pd.isnull(cell_val) else cell_val)

df["is_train"] = np.random.randint(100, size = len(df)) <=70

df_tr = df[df.is_train == True]
df_tr = df_tr.drop('is_train', 1)
df_tr.index = range(len(df_tr))

df_tr_x = df_tr.drop('NDF', 1)


df_tst = df[df.is_train == False]
df_tst = df_tst.drop('is_train', 1)
df_tst.index = range(len(df_tst))

df_tst_x = df_tst.drop('NDF', 1)

print len(df_tr)
print len(df_tst)
```

```python
from sklearn import grid_search
from sklearn.svm import SVC
from sklearn.feature_extraction import DictVectorizer
from __future__ import division


dv = DictVectorizer()
dv.fit(df_tr.T.to_dict().values())

s = dv.transform(df_tr_x.T.to_dict().values())
 #   print s

LR = SVC( degree=3, gamma=0.0, coef0=0.0, shrinking=True, probability=False, tol=0.001, cache_size=200, verbose=False, max_iter=-1, random_state=None)

parameters = {'kernel':['linear', 'poly', 'rbf', 'sigmoid'], 'C':(0.01,1,5), 'class_weight': (None, 'auto'), 'random_state': [1,9]}

clf = grid_search.GridSearchCV( LR, parameters)
clf.fit(s, df_tr.NDF)
    
    
```

```python
print clf.best_estimator_
clf.grid_scores_
```

```python
'REASON_TYPE=Ship Damage'.split('=')

feature_cnt = {}
for feature in dv.feature_names_:
    if '=' in feature:
        col = feature.split('=')[0]
        col_val = feature.split('=')[1]
        feature_cnt[feature] = len(df[df[col] == col_val])
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
import numpy 

c = dv.transform(df_tst_x.T.to_dict().values())
y_true, y_pred = df_tst.NDF, clf.predict(c)

print(classification_report(numpy.asarray(y_true, dtype="|S600"), numpy.asarray(pd.Series(y_pred), dtype="|S600")))

print "\nOVERALL ACCURACY ::\t", accuracy_score(y_true, y_pred)
```
